```python
import sys
from pathlib import Path

sys.path.append(str(Path().resolve().parent))
```

```python
import pandas as pd
from src.paths import TRASNFORMED_DATA_DIR

df = pd.read_parquet(TRASNFORMED_DATA_DIR / 'tabular_data.parquet')
df
```

```python
from datetime import datetime
from src.data_split import train_test_split

X_train, y_train, X_test, y_test = train_test_split(
    df=df,
    cutoff_data=datetime(2025,6,1,0,0,0),
    target_colum_name='target_rides_next_hour'
)

print(f'{X_train.shape=}')
print(f'{y_train.shape=}')
print(f'{X_test.shape=}')
print(f'{y_test.shape=}')
```

```python
import xgboost as xgb
```

```python
# Use only past rides data
past_rides_colimns = [ c for c in X_train.columns if c.startswith('rides_') ]
X_train_only_numeric = X_train[past_rides_colimns]
```

```python
model = xgb.XGBRegressor()
model.fit(X_train_only_numeric, y_train)
```

```python
X_test_only_numeric = X_test[past_rides_colimns]
predictions = model.predict(X_test_only_numeric)
predictions
```

```python
from sklearn.metrics import mean_absolute_error
test_mae = mean_absolute_error(y_test, predictions)
print(f'{test_mae=:.4f}')
```
